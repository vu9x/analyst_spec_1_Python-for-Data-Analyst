1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
import time
from tqdm import tqdm
tqdm.pandas()

def delta_renovated(x):
    time.sleep(0.0001)
    if (x < 0):
        return 0
    
    return x


df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df['delta_renovated'] = df['delta_renovated'].progress_apply(delta_renovated)
df['delta_renovated']
df.head(5)
```

# Условие 3: 3 задача

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
bedrooms = df.groupby('bedrooms')
bedrooms.agg({'price': 'mean'}).sort_values(by='price', ascending=True)
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.pivot_table(index='condition', aggfunc={'price': ['min', 'mean', 'max']})
```

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index=['waterfront', 'condition'], aggfunc={'id': 'count'}).rename(columns={'id': 'house numbers'})
```

3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
table4 = df.pivot_table(index=['floors', 'bedrooms'], aggfunc={'id': 'count'}).rename(columns={'id': 'house_numbers'})
table4[table4.house_numbers == table4.house_numbers.max()]
```

3.5 Постройте таблицу домов в данных в зависимости от состояния дома и оценки дома с подсчетом медианной стоимости

```python
table5 = df.pivot_table(index=['condition', 'grade'], aggfunc={'price': 'median'}).rename(columns={'price': 'median_price'})
table5
```
