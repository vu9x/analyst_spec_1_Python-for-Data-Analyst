# Условие 2: 2 задача

2.1. Изучите распределение домов от наличия вида на набережную

```python
waterfront = house_df['waterfront'].value_counts()

plt.pie(waterfront.values, autopct='%1.1f%%')
plt.legend(waterfront.index)
plt.title('Доля домов с видом на море')
```

Подавляющее кол-во домов не имеют вид на море

2.2. Изучите распределение этажей домов

```python
floors = house_df['floors'].value_counts()

plt.pie(floors.values, autopct='%1.1f%%')
plt.legend(floors.index)
plt.title('Доля домов по их этажности')
```

Почти 90% домов на рынке имеют 1 или 2 этажа

2.3. Изучите распределение состояния домов

```python
condition = house_df['condition'].value_counts()
plt.bar(condition.index, condition.values)
plt.xlabel('condition')
plt.ylabel('Кол-во домов')
plt.title('Распределение домов по состоянию');
```

Предложение ветхих домов на рынке минимально. Состояние большинство домов на рынке представлены в удовлетворительном состоянии. Также немало предложений хороших и находящихся в отличном состоянии домов

# Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
house_df.head(1)
```

```python
pivot_table1 = np.round(
    pd.pivot_table(
        house_df, values='price', index=['floors'],
        aggfunc=['mean', 'median'], fill_value=0
    )
)
pivot_table1.plot.bar(figsize=(6,4), title='Mean and median house prices by number of floors');
```

```python
pivot_table1 = np.round(
    pd.pivot_table(
        house_df, values='price', index=['condition'],
        aggfunc=['mean', 'median'], fill_value=0
    )
)
pivot_table1.plot.bar(figsize=(6,4), title='Mean and median house prices by condition');
```

```python
pivot_table1 = np.round(
    pd.pivot_table(
        house_df, values='price', index=['grade'],
        aggfunc='mean', fill_value=0
    )
)
pivot_table1.plot.bar(figsize=(6,4), title='Mean house price by house grade');
```

```python
pivot_table1 = np.round(
    pd.pivot_table(
        house_df, values='price', index=['waterfront'],
        aggfunc=['mean', 'median', 'max'], fill_value=0
    )
)
pivot_table1.plot.bar(figsize=(6,4), title='Mean, median, max house prices by house grade');
```

1. В среднем, чем больше этажей, тем более высокие цены дома. На графике 'Mean and median house prices by number of floors' видно что разница между медианой и средней больше всего у домов с 2.5 и 3.5 этажей, это означает то, что на рынке у этих домов наблюдаются аномалий в предложениях с высокими ценами
2. В среднем тем лучше состояние дома, тем выше стоимость
3. Чем выше класс дома, тем выше стоимость
4. В среднем дома с видом на побережие моря стоят дороже домов без вида на море. Но среди максимальных цен, есть предложение дома с видом на материк выше чем с видом на море
